```python
import ltspice
import matplotlib.pyplot as plt
filepath = '20210129_nociceptor.raw'
l = ltspice.Ltspice(filepath)
l.parse() # Data loading sequence. It may take few minutes for huge file.

time = l.get_time()
V1 = l.get_data('V(gm)')
```

```python
# %matplotlib widget
plt.figure(figsize=(14,7))
item = -10
plt.plot(time[:item],V1[:item],'-o')
plt.show()
```
